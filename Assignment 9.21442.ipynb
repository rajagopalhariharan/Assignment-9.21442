```python
#Exercise 1
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```python
fig = plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
#Exercise 3
fig = plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title("Zoom")
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```
